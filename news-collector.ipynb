```python
import newspaper
import os
import random
import pickle

url="https://bdnews24.com/"
paper_today = newspaper.build(url,  memoize_articles=False)

cwd=os.getcwd()+'/'
newspaper_path='news_today.pkl'
```

```python
class paragraph:    
    def __init__(self, title, url, text):
        self.title=title
        self.url=url
        self.body=text
```

```python
class customizedPaper:
    def __init__(self):
        self.paragraphs=[]
    def addParagraph(self, para):
        self.paragraphs.append(para)
```

```python
cnt=0
for article in paper_today.articles:
    try:
        article.download()
        article.parse()
        cnt+=1
        print(str(100*cnt/len(paper_today.articles))+"%")
    except:
        article=None
```

```python
with open(newspaper_path, 'wb') as output:
    pickle.dump(paper_today, output, pickle.HIGHEST_PROTOCOL)
```

```python
with open(newspaper_path, 'rb') as input:
    paper_today = pickle.load(input)
```

```python
cnt=0
for mmm in paper_today.articles:
    if mmm is not None:
        print(mmm.title)
        cnt=cnt+1
print(cnt)
```
